```python
# --- 1. DADOS INICIAIS ---
vendas_agosto = {
    "04/08/2025": 2350, "05/08/2025": 1980, "06/08/2025": 2540, "07/08/2025": 2100, "08/08/2025": 3050,
    "11/08/2025": 2650, "12/08/2025": 2180, "13/08/2025": 2760, "14/08/2025": 2920, "15/08/2025": 3100,
    "18/08/2025": 2300, "19/08/2025": 2430, "20/08/2025": 2210, "21/08/2025": 2790, "22/08/2025": 2970,
    "25/08/2025": 3100, "26/08/2025": 2880, "27/08/2025": 2600, "28/08/2025": 2500, "29/08/2025": 3400
}

# --- 2. PREPARAÇÃO DOS DADOS ---
vendas_agosto_valores = list(vendas_agosto.values())
vendas_agosto_keys = list(vendas_agosto.keys())

# --- 3. CÁLCULOS GERAIS DO MÊS ---
vendas_agosto_soma = sum(vendas_agosto_valores)
vendas_agosto_media = vendas_agosto_soma / len(vendas_agosto)

# --- 4. SEPARAÇÃO DAS VENDAS POR SEMANA ---
semana_1 = vendas_agosto_valores[0:5]
semana_2 = vendas_agosto_valores[5:10]
semana_3 = vendas_agosto_valores[10:15]
semana_4 = vendas_agosto_valores[15:20]

# --- 6. CÁLCULO DAS SOMAS SEMANAIS ---
semana_total_1 = sum(list(semana_1))
semana_total_2 = sum(list(semana_2))
semana_total_3 = sum(list(semana_3))
semana_total_4 = sum(list(semana_4))

# --- 7. CÁLCULO DAS MÉDIAS SEMANAIS ---
semana_total_1_media = semana_total_1 / len(semana_1)
semana_total_2_media = semana_total_2 / len(semana_2)
semana_total_3_media = semana_total_3 / len(semana_3)
semana_total_4_media = semana_total_4 / len(semana_4)

# --- 8. ANÁLISE DE MAIOR E MENOR VENDA ---
dia_maior_venda = max(vendas_agosto_valores)
index_maior_venda = vendas_agosto_valores.index(dia_maior_venda)
data_maior_venda = vendas_agosto_keys[index_maior_venda]
dia_minimo_venda = min(vendas_agosto_valores)
index_minimo_venda = vendas_agosto_valores.index(dia_minimo_venda)
data_minimo_venda = vendas_agosto_keys[index_minimo_venda]

# --- 9. CÁLCULO PERCENTUAL ---
diferencial_percentual = ((dia_maior_venda - vendas_agosto_media) / dia_maior_venda) * 100

# --- RELATÓRIO FINAL ---
# Novo design utilizando espaços para alinhamento e caracteres | e -
print("\n+--------------------------------------------------+")
print("|        RELATÓRIO DE VENDAS - AGOSTO 2025         |")
print("+--------------------------------------------------+")

# Seção de Análise Geral
print("\n[ ANÁLISE GERAL DO MÊS ]")
print("-" * 50)
# Usamos f-string com padding para alinhar o texto
print(f"{'Faturamento Total':<25}: R$ {vendas_agosto_soma:,.2f}")
print(f"{'Média de Vendas Diárias':<25}: R$ {vendas_agosto_media:,.2f}")

# Seção de Desempenho Semanal
print("\n[ DESEMPENHO SEMANAL ]")
print("-" * 50)
print(f"  - Semana 1: R$ {semana_total_1:,.2f} (Média: R$ {semana_total_1_media:,.2f})")
print(f"  - Semana 2: R$ {semana_total_2:,.2f} (Média: R$ {semana_total_2_media:,.2f})")
print(f"  - Semana 3: R$ {semana_total_3:,.2f} (Média: R$ {semana_total_3_media:,.2f})")
print(f"  - Semana 4: R$ {semana_total_4:,.2f} (Média: R$ {semana_total_4_media:,.2f})")

# Seção de Destaques
print("\n[ DESTAQUES DO MÊS ]")
print("-" * 50)
print(f"{'Melhor Dia de Vendas':<25}: {data_maior_venda} (R$ {dia_maior_venda:,.2f})")
print(f"{'Pior Dia de Vendas':<25}: {data_minimo_venda} (R$ {dia_minimo_venda:,.2f})")
print(f"{'Pico de Vendas vs Média':<25}: +{diferencial_percentual:.2f}%")

print("\n" + "-"*50)
```
